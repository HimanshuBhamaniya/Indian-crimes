```python
df['Date Reported']= pd.to_datetime(df['Date Reported'], format="%d-%m-%Y %H:%M")
df['Date Case Closed']= pd.to_datetime(df['Date Case Closed'], format="%d-%m-%Y %H:%M")
```

```python
df['Days Took To Close']= (df['Date Case Closed']-df['Date Reported']).dt.days
```

```python
df.head(10)
```

```python
Avg_days_took_close = df['Days Took To Close'].mean()
Avg_Days_Took_to_Close = np.round(Avg_days_took_close, 2)
Avg_Days_Took_to_Close 
```

It is the average days took to close cases.

```python
df.groupby('Crime Description')['Days Took To Close'].mean().sort_values(ascending=False)
```
