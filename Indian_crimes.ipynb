```python
df['Time of Occurrence']= pd.to_datetime(df['Time of Occurrence'], format="%d-%m-%Y %H:%M")
```

```python
df['Only_Time_of_Occurrence'] = df['Time of Occurrence'].dt.time
```

```python
df['Only_Hours_of_Occurrence'] = df['Time of Occurrence'].dt.hour
```

```python
bins=[0,4,8,12,16,20,24]
labels=['Midnight','Dawn','Morning','Afternoon','Evening','Night']
df['Time_period'] = pd.cut(df['Only_Hours_of_Occurrence'], bins=bins, labels=labels, right=False)
```

```python
df.groupby('Time_period', observed=False)['Crime Description'].value_counts().groupby(level=0, observed=False).head(1)
```

Here we can see what sort crimes are happening at what period of time.

```python
df.head()
```

```python
ages= [0,18,20,50,80]
group= ['Under-18','Teenager','Adult','Elder']
df['Victim_Age_Group']= pd.cut(df['Victim Age'], bins=ages, labels=group, right=False)
```

```python
df['Victim_Age_Group'].value_counts()
```

```python
df['Victim_Age_Group'].value_counts(normalize=True)*100
```
